```python
def iqr_outlier_counts(series):
    q1 = series.quantile(0.25); q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return ((series < lower) | (series > upper)).sum()

outlier_counts = {c: iqr_outlier_counts(df[c]) for c in num_cols}
pd.Series(outlier_counts).to_frame("iqr_outlier_count").to_csv("outputs/outlier_counts.csv")
outlier_counts
# show extreme percentiles
percentiles = df[num_cols].quantile([0.001, 0.01, 0.05, 0.5, 0.95, 0.99, 0.999]).T
percentiles.to_csv("outputs/percentiles.csv")
percentiles
```

```python
class OutlierClipper(BaseEstimator, TransformerMixin):
    """Clip features by lower/upper quantiles per-column (works on numpy arrays)."""
    def __init__(self, lower_quantile=0.01, upper_quantile=0.99):
        self.lower_quantile = lower_quantile
        self.upper_quantile = upper_quantile
        self.lower_bounds_ = None
        self.upper_bounds_ = None
    
    def fit(self, X, y=None):
        X = np.asarray(X, dtype=float)
        # compute per-column quantiles
        self.lower_bounds_ = np.quantile(X, self.lower_quantile, axis=0)
        self.upper_bounds_ = np.quantile(X, self.upper_quantile, axis=0)
        return self
    
    def transform(self, X):
        X = np.asarray(X, dtype=float).copy()
        # clip per column
        X = np.clip(X, self.lower_bounds_, self.upper_bounds_)
        return X
```

```python
# Prepare X
X = df.drop(columns=['label']).copy()
y = df['label'].copy()

clipper = OutlierClipper(lower_quantile=0.01, upper_quantile=0.99)
clipper.fit(X.values)

# before: boxplot
plt.figure(figsize=(12,5))
X.boxplot(rot=45)
plt.title("Before clipping")
plt.tight_layout()
plt.savefig("outputs/plots/before_clipping_boxplot.png")
plt.close()

# after: apply clipper and plot
X_clipped = pd.DataFrame(clipper.transform(X.values), columns=X.columns)
plt.figure(figsize=(12,5))
X_clipped.boxplot(rot=45)
plt.title("After clipping (1%-99%)")
plt.tight_layout()
plt.savefig("outputs/plots/after_clipping_boxplot.png")
plt.close()
```

```python
numeric_features = X.columns.tolist()
numeric_transformer = Pipeline(steps=[
    ("clipper", OutlierClipper(lower_quantile=0.01, upper_quantile=0.99)),
    ("scaler", StandardScaler())
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features)
    ], remainder='drop'  # drop other columns
)
```

```python
le = LabelEncoder()
y_enc = le.fit_transform(y)  # label encoder saved later

X_train, X_test, y_train, y_test = train_test_split(
    X, y_enc, test_size=0.2, stratify=y_enc, random_state=RANDOM_STATE
)
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
```

```python
rf_pipeline = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=-1))
])

xgb_pipeline = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", XGBClassifier(use_label_encoder=False, eval_metric="mlogloss", random_state=RANDOM_STATE, n_jobs=-1))
])
```

```python
# cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=RANDOM_STATE)

# print("RF CV (accuracy):", cross_val_score(rf_pipeline, X, y_enc, cv=cv, scoring='accuracy', n_jobs=-1).mean())
# print("XGB CV (accuracy):", cross_val_score(xgb_pipeline, X, y_enc, cv=cv, scoring='accuracy', n_jobs=-1).mean())
rf_pipeline.fit(X_train, y_train)
rf_acc = rf_pipeline.score(X_test, y_test)
print("RF Accuracy (test split):", rf_acc)
```

```python
def fit_and_eval(pipe, name):
    pipe.fit(X_train, y_train)
    preds = pipe.predict(X_test)
    acc = accuracy_score(y_test, preds)
    f1 = f1_score(y_test, preds, average='weighted')
    print(f"\n{name} Results -> Accuracy: {acc:.4f}, Weighted F1: {f1:.4f}")
    print(classification_report(y_test, preds, target_names=le.classes_))
    return pipe, preds, acc, f1

rf_trained, rf_preds, rf_acc, rf_f1 = fit_and_eval(rf_pipeline, "RandomForest")
xgb_trained, xgb_preds, xgb_acc, xgb_f1 = fit_and_eval(xgb_pipeline, "XGBoost")

# choose best by weighted F1
if rf_f1 >= xgb_f1:
    best_model = rf_trained
    best_name = "RandomForest"
    best_preds = rf_preds
    best_acc = rf_acc
    best_f1 = rf_f1
else:
    best_model = xgb_trained
    best_name = "XGBoost"
    best_preds = xgb_preds
    best_acc = xgb_acc
    best_f1 = xgb_f1

print(f"\nSelected best model: {best_name} (weighted F1: {best_f1:.4f})")
```

```python
cm = confusion_matrix(y_test, best_preds)
plt.figure(figsize=(12,10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title(f"Confusion Matrix - {best_name}")
plt.tight_layout()
plt.savefig(f"outputs/plots/confusion_matrix_{best_name}.png")
plt.close()

cm_norm = confusion_matrix(y_test, best_preds, normalize='true')
plt.figure(figsize=(12,10))
sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title(f"Normalized Confusion Matrix - {best_name}")
plt.tight_layout()
plt.savefig(f"outputs/plots/confusion_matrix_norm_{best_name}.png")
plt.close()
```

```python
model_obj = best_model.named_steps['model']
# feature importances (tree models)
if hasattr(model_obj, "feature_importances_"):
    importances = model_obj.feature_importances_
    feat_names = numeric_features
    imp_df = pd.DataFrame({'feature': feat_names, 'importance': importances}).sort_values('importance', ascending=False)
    plt.figure(figsize=(8,5))
    sns.barplot(x='importance', y='feature', data=imp_df)
    plt.title(f"Feature Importances - {best_name}")
    plt.tight_layout()
    plt.savefig(f"outputs/plots/feature_importance_{best_name}.png")
    plt.close()
else:
    print("Model does not expose feature_importances_")
```

```python
try:
    # preprocessed training data (numpy)
    X_train_pre = best_model.named_steps['preprocess'].transform(X_train)
    # use a small sample for SHAP to keep compute reasonable
    sample_idx = np.random.choice(range(X_train_pre.shape[0]), size=min(500, X_train_pre.shape[0]), replace=False)
    X_shap = X_train_pre[sample_idx]

    explainer = shap.TreeExplainer(best_model.named_steps['model'])
    shap_values = explainer.shap_values(X_shap)  # multiclass -> list of arrays
    # For multiclass, shap_values is list; show summary for the first class as an example:
    plt.figure()
    shap.summary_plot(shap_values[0], X_shap, feature_names=numeric_features, show=False)
    plt.tight_layout()
    plt.savefig("outputs/plots/shap_summary_class0.png", bbox_inches='tight')
    plt.close()
    print("Saved SHAP summary (class 0) plot.")
except Exception as e:
    print("SHAP step skipped or failed:", e)
```

```python
joblib.dump(best_model, f"models/best_model_{best_name}.pkl")
joblib.dump(le, "models/label_encoder.pkl")
print("Saved model and label encoder to models/ folder.")
```

```python
def recommend_crop(model, label_encoder, sample_dict):
    # sample_dict: {'N': val, 'P': val, ... } keys must match numeric_features
    row = pd.DataFrame([sample_dict], columns=numeric_features)
    pred_enc = model.predict(row)[0]
    pred_crop = label_encoder.inverse_transform([pred_enc])[0]
    probs = None
    if hasattr(model.named_steps['model'], 'predict_proba'):
        proba = model.predict_proba(row)[0]
        top_idx = np.argsort(proba)[::-1][:5]
        probs = [(label_encoder.inverse_transform([i])[0], float(proba[i])) for i in top_idx]
    return pred_crop, probs

# Example usage
sample = {'N':90, 'P':42, 'K':43, 'temperature':20, 'humidity':80, 'ph':6.5, 'rainfall':200}
crop, probs = recommend_crop(best_model, le, sample)
print("Recommended crop:", crop)
print("Top probabilities:", probs)
```
